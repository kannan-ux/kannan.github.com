```python
import pandas as pd
```

```python
weather_data=[("google","salesExcetive","bachelors",0),
              ("google","developer","bachelors",1),
              ("google","salesExcetive","masters",1),
              ("google","businessmanager","bachelors",0),]
```

```python
df=pd.DataFrame(weather_data,columns=["comapny","position","degree","salarymorethen100k"])
```

```python
X=df.drop('salarymorethen100k',axis=1)
```

```python
Y=df['salarymorethen100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_comapny=LabelEncoder()
le_position=LabelEncoder()
le_degree=LabelEncoder()
```

```python
X['company_n']=le_comapny.fit_transform(X["comapny"])
```

```python
X['degree_n']=le_degree.fit_transform(X["degree"])
X['position_n']=le_position.fit_transform(X["position"])
```

```python
X2=X.drop(['comapny','position','degree' ],axis='columns')
```
